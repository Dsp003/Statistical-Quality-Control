```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def comb(x,y):
    u = p = 1
    k, t = min(x-y, y), max(x-y, y)
    while(x > t):
        u *= x
        p *= k
        x -= 1
        k -= 1
    return u / p
```

```python
pop = np.random.randint(100, size=(10000))
ls = list()
for i in range(1000):
    ls.append(pop[np.random.choice(10000, size=(100), replace=False)].mean())
plt.figure()
plt.hist(ls, bins=100)
plt.title('Histogram representing the spread of sample means')
plt.ylabel('frequecy')
plt.xlabel('mean of samples')
```

```python
out = pd.DataFrame({'Generated Data': [np.mean(pop), np.var(pop)],
                    'Random Samples of size 100': [np.mean(ls), np.var(ls)],
                    'Predicted by Central Limit Theorem': [np.mean(pop), (np.var(pop)) / 100]
                   }, index=['mean', 'variance'])
out
```
